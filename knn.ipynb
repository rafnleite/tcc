Importando libs necess√°rias

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import time
import numpy as np
```

```python
TRAIN_SIZE = 0.8
```

```python
def grid_search(k, distancia, algoritmo, datasets, repeticoes):
    resultados = []
    for l1 in k:
        for l2 in distancia:
            for l3 in algoritmo:
                for l4 in range(len(datasets)):
                    resultado_parcial = []
                    inicio = time.time()
                    for l5 in range(repeticoes):
                        X=datasets[l4]["dataset"][datasets[l4]["preditoras"]]
                        y=datasets[l4]["dataset"][datasets[l4]["resultado"]]
                        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1-TRAIN_SIZE)
                        knn = KNeighborsClassifier(n_neighbors=l1, metric=l2, weights=l3)
                        knn.fit(X_train, y_train)
                        y_pred = knn.predict(X_test)
                        resultado_parcial.append(accuracy_score(y_test, y_pred))
                    arr = np.array(resultado_parcial)
                    resultado_parcial_metricas = {"k": l1,
                                     "distancia": l2,
                                     "algoritmo": l3,
                                     "tempo": time.time() - inicio,
                                     "precisao_maior": np.max(arr),
                                     "precisao_menor": np.min(arr),
                                     "precisao_media": np.mean(arr),
                                     "precisao_desvio_padrao": np.std(arr),
                                     "precisao_p25": np.percentile(arr,25),
                                     "precisao_p50": np.percentile(arr,50),
                                     "precisao_p75": np.percentile(arr,75)
                                    }
                    print(resultado_parcial_metricas)
                    resultados.append(resultado_parcial_metricas)
                    
    return resultados
```
